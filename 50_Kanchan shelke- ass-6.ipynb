### Build the web crawler to pull product information and links from an e-commerce website.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

### Request page source from url

```python
url = "https://www.imdb.com/chart/top/"
```

```python
HEADERS = {'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148'}
```

```python
page = requests.get(url, headers=HEADERS)
page
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
scraped_movies = soup.find_all('h3', class_='ipc-title__text')
scraped_movies
```

```python
movies=[]
for movie in scraped_movies:
    movie=movie.get_text().replace('\n',"")
    movie=movie.strip(" ")
    movies.append(movie)
movies
```

```python
scraped_ratings = soup.find_all('span', class_="ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating")
scraped_ratings
```

```python
ratings = []
for rating in scraped_ratings:
    rating = rating.get_text().replace('\n','')
    ratings.append(rating)
ratings
```

```python
# as total values in 'ratings' lists are 250 and 'movies' we need to trim last 13 values from movies
movies = movies[: len(movies)-13]
movies
```

### Store the scraped data

```python
data = pd.DataFrame()
data['Movie Names'] = movies
data['Ratings'] = ratings
data.head()
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
def crawl_web(seed_url, max_pages):
    pages_visited = 0
    url_queue = [seed_url]
    
    while pages_visited < max_pages and url_queue:
        url = url_queue.pop(0)
        
        try:
            response = requests.get(url)
            if response.status_code == 200:
                soup = BeautifulSoup(response.text, 'html.parser')
                
                print(soup.title.string)
                
                links = [a['href'] for a in soup.find_all('a', href=True)]
                for link in links:
                    url_queue.append(link)
                    
                pages_visited += 1
        except Exception as e:
            print(f"Error:{e}")
            
       
crawl_web("https://google.com", max_pages=10)
            
```
